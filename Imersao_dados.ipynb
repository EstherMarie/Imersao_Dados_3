```python
# renomeando a coluna 'droga' para 'composto'
mapa = {'droga': 'composto'}
dados.rename(columns=mapa, inplace=True) # dados.rename(columns=mapa)não troca o nome na coluna original. Para isso precisamos utilizar o inplace=True 

dados.head()
```

```python
# compostos = dados['composto'].value_counts()[0:5] # para mostrar os elementos da linha 0 até a 4
# compostos # mostra o código do composto e a quantidade de elementos
```

```python
# compostos.index # mostra apenas o nome dos compostos

cod_compostos = dados['composto'].value_counts().index[0:5]

cod_compostos
```

```python
# query
dados.query('composto in @cod_compostos') # retorna tudo que for igual a 'cod_composto' na coluna 'composto'. É nessessário informar que o 'cod_composto' foi declarado fora da string, para isso utilizamos o @
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set() # para deixar o gráfico no padrão do seaborn
plt.figure(figsize=(8,6)) # tamanho do gráfico no eixo x e y respectivamente
ax = sns.countplot(x = 'composto', order=cod_compostos, data=dados.query('composto in @cod_compostos'))
ax.set_title('Top 5 Compostos') # para definir um título e deixá-lo mais visível
plt.show() # limpa o texto padrão que aparece antes da figura
```
